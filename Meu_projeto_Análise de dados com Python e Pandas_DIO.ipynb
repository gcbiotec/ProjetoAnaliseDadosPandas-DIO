```python
  df_excel["Data"] = df_excel["Data Venda"].astype("int64")
```

```python
df_excel.groupby(df_excel["Data Envio"].dt.year)["Valor Venda"].sum()
```

```python
#Criando uma nova coluna com o ano
df_excel["Ano_Venda"] = df_excel["Data Envio"].dt.year
```

```python
df_excel.sample(3)
```

```python
# Extraindo o mês e o dia das datas:
df_excel["mes_venda"], df_excel["dia_venda"] = (df_excel["Data Venda"].dt.month, df_excel["Data Venda"].dt.day)
```

```python
df_excel.sample(3)
```

```python
# Data mais antiga:
df_excel["Data Envio"].min()
```

```python
# Calculando o tempo entre a compra do produto e o envio:
df_excel['Tempo_compra_envio'] = df_excel['Data Envio'] - df_excel['Data Venda']
```

```python
df_excel.head()
```

```python
#Criando a coluna de trimestre
df_excel["trimestre_venda"] = df_excel["Data Envio"].dt.quarter
```

```python
df_excel.tail()
```

```python
#Filtrando as vendas de 2008 do mês de abril:
vendas_abril_08 = df_excel.loc[(df_excel["Data Envio"].dt.year == 2008) & (df_excel["Data Envio"].dt.month == 4)]
```

```python
vendas_abril_08.sample(15)
```

```python
vendas_abril_08["dia_venda"].value_counts(ascending=False).plot.bar()
```

```python
vendas_abril_08["Valor Venda"].value_counts().plot.barh()
```

```python
vendas_abril_08.groupby(vendas_abril_08["Data Venda"].dt.day)["Valor Venda"].sum().plot.pie()
```

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
vendas_abril_08["Fabricante"].value_counts().plot.bar(title="Total vendas por Fabricante", color="red")
plt.xlabel("Fabricante")
plt.ylabel("Quantidade");
```

```python
vendas_abril_08.groupby(vendas_abril_08["dia_venda"])["Receita/Vendas"].sum().plot(title = "Total Produtos vendidos x mês")
plt.xlabel("Mês")
plt.ylabel("Receita/Vendas")
plt.legend();
```

```python
plt.scatter(x=vendas_abril_08["dia_venda"], y = vendas_abril_08["Receita/Vendas"]);
plt.xlabel("Mês")
plt.ylabel("Receita/Vendas")
plt.legend();
```
